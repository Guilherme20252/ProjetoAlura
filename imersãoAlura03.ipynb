```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
sns.barplot(data=df, x= 'experience_level', y='salary_in_usd')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x= 'experience_level', y='salary_in_usd')
plt.title("salário por nível de experiência")
plt.xlabel("nível de experiência")
plt.ylabel("salário em dólar")
plt.show()
```

```python
df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending= False)
```

```python
ordem = df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending= True).index
ordem
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x= 'experience_level', y='salary_in_usd', order= ordem)
plt.title("salário por nível de experiência")
plt.xlabel("nível de experiência")
plt.ylabel("salário em dólar")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['salary_in_usd'], bins = 50, kde=True )
plt.title("salário por nível de experiência")
plt.xlabel("Salário em usd")
plt.ylabel("Frequencia")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x= df['salary_in_usd'])
plt.title("Boxplot Salário")
plt.xlabel("Salário em USD")
plt.show()
```

```python
# The original 'ordem_senioridade' list is not directly applicable to the 'experience_level' column.
# Using 'ordem' which was previously defined based on 'experience_level' and salary_in_usd in cell Ai1vmIoTXoz7.
plt.figure(figsize=(8,4))
sns.boxplot(x= 'experience_level', y='salary_in_usd', data=df, order= ordem)
plt.title("Boxplot da distribuição por Senioridade")
plt.xlabel("Salário em USD")
plt.show()
```

```python
ordem_senioridade = ['junior', 'pleno', 'senior', 'execultivo']
plt.figure(figsize=(8,4))
sns.boxplot(x= 'experience_level', y='salary_in_usd', data=df, order= ordem, palette='Set2')
plt.title("Boxplot da distribuição por Senioridade")
plt.xlabel("Salário em USD")
plt.show()
```

```python
import plotly.express as px
```

```python
df_grouped = df.groupby('experience_level')['salary_in_usd'].mean().reindex(ordem)
fig = px.bar(
    x=df_grouped.index,
    y=df_grouped.values,
    title='Média Salarial por Nível de Experiência',
    labels={'x': 'Nível de Experiência', 'y': 'Salário Médio em USD'}
)
fig.show()
```
